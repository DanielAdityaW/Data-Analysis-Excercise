```python
kind = df.groupby(['category'])['product_id'].count()

plt.bar(kind.index, kind, color='#ff1a75', width=0.5)

plt.ylabel("Jumlah masing - masing product", color='#ff1a75',size='10')
plt.xlabel("jenis produk yang ada", color='#ff1a75')
plt.title("Jumlah produk yang ada berdasarkan kategori", color='#ff1a75', size='10')
plt.savefig('barchart.png')
files.download('barchart.png')
plt.show()
```

```python
kind = df.groupby(['category'])['product_id'].count()
plt.plot(kind.index, kind, color='red')
plt.grid(color='green', linestyle='--', linewidth=2, axis='y', alpha=0.2)
plt.ylabel("Jumlah masing - masing product", color='#336600',size='10')
plt.xlabel("jenis produk yang ada", color='#336600')
plt.title("Jumlah produk yang ada berdasarkan kategori", color='#336600', size='10')
plt.show()
```

```python
nilai = df.groupby(['rating'])['category'].count().iloc[:7]
kat = df.groupby(['category'])['product_id'].count().iloc[:7]

plt.bar(nilai.index, nilai, color='green', width=0.5)

plt.grid(color='green', linestyle='--', linewidth=2, axis='y', alpha=0.2)
plt.ylabel("Rating", color='green')
plt.xlabel("Kategori Produk", color='green')
plt.title("Rating Produk Berdasarkan Kategori", color='green')
plt.show()
```

```python
Jumlahprodukterjual=df.groupby(['product_name'])['sold'].count().iloc[:15]

plt.barh(Jumlahprodukterjual.index, Jumlahprodukterjual) #barhuntukhorizontal
plt.ylabel("Nama Produk")
plt.xlabel("Jumlah")
plt.title("Jumlah produk terjual pada 15 data pertama")
plt.savefig('vertical_bar.png')
files.download('vertical_bar.png')
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
tips['tip_pct']=200*tips['tip']/tips['total_bill']

grid =sns.FacetGrid(tips,row="sex", col="day", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
plt.savefig('chart.png')
files.download('chart.png')
```

```python
tips['tip_pct']=100*tips['tip']/tips['total_bill']

grid =sns.FacetGrid(tips,row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)
```

```python
with sns.axes_style(style='ticks'):
  x = sns.catplot("sex", "total_bill", "smoker", data=tips, kind="box")
  x.set_axis_labels("Sex", "Total Bill");
  plt.savefig('enigmacampboxplot.jpg')
  files.download('enigmacampboxplot.jpg')
```
